## PURCHASE
### Including updating inventory and sales json

```python
sales={}
```

```python
print('\t\t\tAZ BOOKSTORE')
print('*******************************************************')
ui_prod  = input("Enter the product_Id: ")
ui_quant = int(input("Enter the quantity: "))
print('*******************************************************')
print('Book\t\t\t\tQty\tRate\tAmount')
print('*******************************************************\n')
print(records[ui_prod]['Book_Name'],'\t',ui_quant,'\t',records[ui_prod]['Price'],'\t',records[ui_prod]['Price'] * ui_quant)
if(ui_quant > records[ui_prod]['Quantity']):
  print(ui_quant,' not in stock. Only ',records[ui_prod]['Quantity'],' available!')
print('\n*******************************************************')
print("Total Amount:\t\t\t\t\t", records[ui_prod]['Price'] * ui_quant)
print('*******************************************************')
print('\nTHANK YOU! COME AGAIN!')
r=len(sales)
s= r+1


records[ui_prod]['Quantity'] = records[ui_prod]['Quantity'] - ui_quant
js = json.dumps(records)


#updating inventory

fd = open("record.json",'w')  
fd.write(js)
fd.close()


#Sales json

sales[s]={'Book_id' : ui_prod, 'Genre':records[ui_prod]['Genre'], 'Qnty' : ui_quant, 'Amount': records[ui_prod]['Price'] * ui_quant}
sale = json.dumps(sales)
fd = open("sale.json",'a')
fd.write(sale)
fd.close()
```
